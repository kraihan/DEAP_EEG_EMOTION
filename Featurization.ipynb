# Wavelet energy

```python
import numpy as np
import os
import pickle
import pywt

# default parameters
subjects = 32
trials = 40
Channel = 32
Fs = 128
Time = 63

# set parameters
frameNum = 60
totalScale = 64
exScale = 32  # This defines the range, but we'll adjust based on actual slice size
wname = 'morl'  # Use a continuous wavelet like 'morl'

# Set path to DEAP data folder
data_path = 'data_preprocessed_python'
output = 'WE'
# Function to load .dat files
def load_dat_file(filename):
    with open(filename, 'rb') as f:
        data = pickle.load(f, encoding='latin1')  # Use latin1 to handle any encoding issues
    return data

# Loop over subject
for subject in range(1, subjects + 1):
    print(f'\nWorking on file number {subject}:\n')
    
    if subject < 10:
        myfilename = f's0{subject}.dat'
    else:
        myfilename = f's{subject}.dat'
    
    # Load data for the subject
    subject_data = load_dat_file(os.path.join(data_path, myfilename))
    data = subject_data['data']
    
    for trial in range(trials):
        print(f'\nCreating file subject {subject}, trial {trial + 1}:\n')
        output_filename = f'WE/subject_{subject}_trial_{trial + 1}.csv'
        
        # Open file for writing output
        with open(output_filename, 'w') as fid:
            output = np.zeros((Channel, totalScale, frameNum))
            datastart = 128 * 3
            datalength = 8064 - 128 * 3

            # Loop over channels
            for channel in range(32):
                data1 = np.zeros(datalength)
                
                for ii in range(datalength):
                    data1[ii] = data[trial, channel, ii + datastart]

                # Decompose into wavelets
                f = np.arange(1, totalScale + 1)
                f = Fs / totalScale * f
                wcf = pywt.central_frequency(wname)
                scal = Fs * wcf / f
                
                # Continuous Wavelet Transform
                coefs, freqs = pywt.cwt(data1, scales=np.arange(1, totalScale + 1), wavelet=wname)

                # Compute wavelet scalogram (energy density)
                sc = np.abs(coefs)**2
                
                frameSize = 60 / frameNum
                start = 0
                
                # Split into frames
                for k in range(frameNum):
                    output[channel, :, k] = np.sum(sc[:, int(start):int(frameSize * (k + 1))], axis=1)
                    start += frameSize
                
                print('.', end='')

            # Extract relevant scales (scales 8 to 39)
            exoutput = output[:, 7:39, :]  # This will have shape (32, 31, 60)

            # Write to file directly (iterate over 31 scales instead of exScale which was set to 32)
            for channel in range(Channel):
                for scale in range(exoutput.shape[1]):  # use exoutput.shape[1] instead of exScale (which is 31)
                    for frame in range(frameNum):
                        fid.write(f'{exoutput[channel, scale, frame]},')
                    fid.write('\n')

        print(f'Saved output to {output_filename}')
        break #remove this line for feature extraction
    break #remove this line for feature extraction
```

# Phase Locking Value 

```python
import numpy as np
import os
import pickle
from scipy.signal import hilbert
from scipy.io import savemat

# Default parameters
Participant = 32
trial = 40
Channel = 32
Fs = 128
Time = 63

# Set parameters
frameNum = 60
exScale = 32  # For output reshaping, assuming PLV results between pairs

# Set path to DEAP data folder
data_path = 'data_preprocessed_python'
output_path = 'PLV'

if not os.path.exists(output_path):
    os.makedirs(output_path)

# Function to load .dat files
def load_dat_file(filename):
    with open(filename, 'rb') as f:
        data = pickle.load(f, encoding='latin1')  # Use latin1 to handle any encoding issues
    return data

# Loop over participants
for participant in range(1, Participant + 1):
    print(f'\nWorking on file number {participant}:\n')
    
    if participant < 10:
        myfilename = f's0{participant}.dat'
    else:
        myfilename = f's{participant}.dat'
    
    # Load data for the participant
    participant_data = load_dat_file(os.path.join(data_path, myfilename))
    data = participant_data['data']
    
    # Loop over videos
    for trial in range(trials):
        print(f'\nCreating file subject {participant}, trial {trial + 1}:\n')
        
        output_filename = f'subject_{participant}_trial_{trial + 1}.csv'
        output_filepath = os.path.join(output_path, output_filename)
        
        # Open file for writing output
        with open(output_filepath, 'w') as fid:
            output = np.zeros((Channel, Channel, frameNum))  # For storing PLV between channel pairs
            datastart = 128 * 3
            datalength = 8064 - 128 * 3
            
            # Prepare data for all channels
            data_segments = np.zeros((Channel, datalength))
            for channel in range(Channel):
                data_segments[channel, :] = data[trial, channel, datastart:datastart + datalength]
            
            # Decompose using the Hilbert transform to get phase information
            phases = np.angle(hilbert(data_segments.T))  # Hilbert transform and phase extraction
            
            frameSize = datalength // frameNum
            start = 0
            
            # Compute PLV for each frame
            for k in range(frameNum):
                end_frame = start + frameSize
                if end_frame > datalength:
                    end_frame = datalength
                
                # Extract phase data for the current frame
                phase_frame = phases[start:end_frame, :]
                N = phase_frame.shape[0]  # Number of time samples in this frame
                
                # Compute PLV between each pair of channels
                for ch1 in range(Channel):
                    for ch2 in range(Channel):
                        phase_diff = phase_frame[:, ch1] - phase_frame[:, ch2]
                        plv_value = np.abs(np.sum(np.exp(1j * phase_diff)) / N)  # PLV calculation
                        output[ch1, ch2, k] = plv_value
                
                start = end_frame
                print('.', end='')
            
            # Extract desired range and reshape
            exoutput = output[:, :exScale, :]
            output2d = np.reshape(exoutput, (Channel * exScale, frameNum))
            
            # Write to file
            for row in output2d:
                fid.write(','.join(map(str, row)) + '\n')
        
        print(f'Saved output to {output_filepath}')
        break
    break
```

# Cross Corelation

```python
import numpy as np
import os
import pickle
from scipy.stats import pearsonr

# Default parameters
Participant = 32
Video = 40
Channel = 32
Fs = 128
Time = 63

# Set parameters
frameNum = 60
exScale = 32  # For output reshaping, assuming cross-correlation results between pairs

# Set path to DEAP data folder
data_path = 'data_preprocessed_python'
output_path = 'XCOR'

if not os.path.exists(output_path):
    os.makedirs(output_path)

# Function to load .dat files
def load_dat_file(filename):
    with open(filename, 'rb') as f:
        data = pickle.load(f, encoding='latin1')  # Use latin1 to handle any encoding issues
    return data

# Loop over participants
for participant in range(1, Participant + 1):
    print(f'\nWorking on file number {participant}:\n')
    
    if participant < 10:
        myfilename = f's0{participant}.dat'
    else:
        myfilename = f's{participant}.dat'
    
    # Load data for the participant
    participant_data = load_dat_file(os.path.join(data_path, myfilename))
    data = participant_data['data']
    
    # Loop over videos
    for video in range(Video):
        print(f'\nCreating file subject {participant}, trial {video + 1}:\n')
        
        output_filename = f'subject_{participant}_trial_{video + 1}.csv'
        output_filepath = os.path.join(output_path, output_filename)
        
        # Open file for writing output
        with open(output_filepath, 'w') as fid:
            output = np.zeros((Channel, Channel, frameNum))  # For storing cross-correlation between channel pairs
            datastart = 128 * 3
            datalength = 8064 - 128 * 3
            
            # Prepare data for all channels
            data_segments = np.zeros((Channel, datalength))
            for channel in range(Channel):
                data_segments[channel, :] = data[video, channel, datastart:datastart + datalength]
            
            frameSize = datalength // frameNum
            start = 0
            
            # Compute cross-correlation for each frame
            for k in range(frameNum):
                end_frame = start + frameSize
                if end_frame > datalength:
                    end_frame = datalength
                
                # Extract data for the current frame
                frame_data = data_segments[:, start:end_frame].T
                
                # Compute cross-correlation for each pair of channels
                for ch1 in range(Channel):
                    for ch2 in range(Channel):
                        corr_value, _ = pearsonr(frame_data[:, ch1], frame_data[:, ch2])  # Pearson correlation coefficient
                        output[ch1, ch2, k] = corr_value
                
                start = end_frame
                print('.', end='')
            
            # Extract desired range and reshape
            exoutput = output[:, :exScale, :]
            output2d = np.reshape(exoutput, (Channel * exScale, frameNum))
            
            # Write to file
            for row in output2d:
                fid.write(','.join(map(str, row)) + '\n')
        
        print(f'Saved output to {output_filepath}')
        break
    break
```

# Diffrential Entropy 

```python
import numpy as np
import os
import pickle
import pywt

# Default parameters
Participant = 32
Video = 40
Channel = 32
Fs = 128  # Sampling frequency
Time = 63

# Set parameters
frameNum = 60  # Number of frames per trial
totalScale = 64  # Number of wavelet scales
exScale = 32  # Extracted range of scales
wname = 'morl'  # Wavelet type: Morlet

# Path to DEAP data folder and DE output folder
data_path = 'data_preprocessed_python'
output_folder = 'DE'

# Create the DE folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Small constant to avoid log(0)
EPSILON = 1e-6

# Function to load .dat files
def load_dat_file(filename):
    with open(filename, 'rb') as f:
        data = pickle.load(f, encoding='latin1')  # Using latin1 encoding to handle compatibility
    return data

# Loop over participants
for participant in range(1, Participant + 1):
    print(f'\nProcessing participant {participant}:\n')

    # File naming
    myfilename = f's{participant:02d}.dat'

    # Load participant data
    participant_data = load_dat_file(os.path.join(data_path, myfilename))
    data = participant_data['data']  # Shape: (40 videos, 32 channels, 8064 samples)

    # Define EEG data segment after removing baseline
    datastart = 3 * Fs  # First 3 seconds (3  128 samples) are the baseline
    datalength = 8064 - datastart  # Remaining data (60 seconds  128 Hz)

    # Correct **frame size in samples**
    frameSize = datalength // frameNum  # Each frame gets equal samples

    # Loop over videos
    for video in range(Video):
        print(f'Processing subject {participant}, trial {video + 1}...\n')

        # Define output filename
        output_filename = os.path.join(output_folder, f'subject_{participant}_trial_{video + 1}.csv')

        # Prepare output array
        output = np.zeros((Channel, totalScale, frameNum))

        # Loop over channels
        for channel in range(Channel):
            # Extract EEG signal for this channel
            data1 = data[video, channel, datastart:datastart + datalength]

            # Compute CWT
            scales = np.arange(1, totalScale + 1)
            coefs, _ = pywt.cwt(data1, scales=scales, wavelet=wname)

            # Compute variance over different frames
            for k in range(frameNum):
                start_idx = k * frameSize
                end_idx = start_idx + frameSize

                if end_idx > coefs.shape[1]:  # Ensure frame doesn't exceed signal length
                    end_idx = coefs.shape[1]

                segment = coefs[:, start_idx:end_idx]

                # Compute variance along time (axis=1)
                var_segment = np.var(segment, axis=1)

                # Normalize variance to avoid near-zero log values
                var_segment = np.maximum(var_segment, EPSILON)

                # Compute Differential Entropy (DE)
                output[channel, :, k] = 0.5 * np.log(2 * np.pi * np.e * var_segment)

            print('.', end='', flush=True)  # Progress indicator

        print("\nExtracting scales 8 to 39...")

        # Extract relevant scales (8 to 39)
        exoutput = output[:, 7:39, :]  # Shape (32, 32, 60)

        # Reshape to 2D: (Channel * exScale, frameNum)
        output2d = exoutput.reshape(Channel * exScale, frameNum)

        # Save output to CSV
        np.savetxt(output_filename, output2d, delimiter=',')

        print(f'Saved DE features to {output_filename}\n')

        break  # Remove this break if you want to process all videos
    break  # Remove this break if you want to process all participants
```
